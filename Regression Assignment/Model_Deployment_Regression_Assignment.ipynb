```python
# Importing necessary libraraies
import pickle
import pandas as pd
```

```python
model = pickle.load(open("Regression_Assignment_Best_Model",'rb'))
```

```python
# User inputs
age = int(input("Enter your age: "))
bmi = float(input("Enter your bmi: "))
children = int(input("Enter no. of children: "))
sex = int(input("enter sex male=1,female=0: "))
smoker = int(input("if smoker enter 1 else enter 0: "))
```

```python
user_data = pd.DataFrame([[age,bmi,children,sex,smoker]],columns=['age', 'bmi', 'children', 'sex_male', 'smoker_yes'])
result = model.predict(user_data)
result
```
