```python
import pickle
```

```python
# Loading the saved model
model = pickle.load(open("SVR_PreProcessed_Input_Saved_Model.sav",'rb'))
sc = pickle.load(open("X_Scaler.pkl",'rb'))
```

```python
# User inputs
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=float(input("Sex Male 0 or 1:"))
smoker_yes_input=float(input("Smoker yes 0 or 1:"))
```

```python
scaled_user_input = sc.transform([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
scaled_user_input
```

```python
result = model.predict(scaled_user_input)
print("The predicted amount: ",result)
```
