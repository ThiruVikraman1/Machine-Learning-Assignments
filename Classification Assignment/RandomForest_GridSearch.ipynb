```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# importing dataset
dataset = pd.read_csv("CKD.csv")
dataset.head()
```

```python
# Categorical data handling
dataset = pd.get_dummies(dataset,drop_first=True,dtype=int)
dataset
```

```python
# Dataset columns
dataset.columns
```

```python
dataset["classification_yes"].value_counts()
```

```python
# Feature and target selection
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent = dataset['classification_yes']
```

```python
independent
```

```python
dependent
```

```python
# Splitting data into training and testing sets
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.25,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# Data standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# GridSearch CV
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Defining the different parameters for gridsearch optimization
param_grid = {'criterion':['gini','entropy','log_loss'],
              'n_estimators':[10,25,50,75,100],
              'class_weight':['balanced','balanced_subsample'],
              'max_features':['sqrt','log2']}


grid = GridSearchCV(RandomForestClassifier(),param_grid,refit=True,verbose=3,scoring='f1_weighted',n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
# Best gridseach parameter
grid.best_params_
```

```python
# Model prediction
y_pred = grid.predict(x_test)
y_pred
```

```python
# Evaluation metrics
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
# Classification report
from sklearn.metrics import classification_report
report = classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```

```python
# Grid results
table = pd.DataFrame(grid.cv_results_)
table
```
