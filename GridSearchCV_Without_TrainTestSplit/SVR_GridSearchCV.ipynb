```python
# Importing necessary libraries
import pandas as pd
import numpy as np
```

```python
# Loading the dataset
dataset = pd.read_csv("insurance_pre.csv")
dataset.head()
```

```python
# Categorical data - One-Hot encodinng
dataset = pd.get_dummies(dataset,drop_first=True,dtype=int)
dataset.head()
```

```python
dataset.columns
```

```python
# Feature selection
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent.head() 
```

```python
# Target or output
dependent = dataset[['charges']]
dependent.head()
```

```python
# GridSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['linear','rbf'], 'C':[1,10,100],'gamma':['auto','scale']}
grid = GridSearchCV(SVR(),param_grid,verbose=3,refit=True,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
# Displaying the best model parameters
print("The best parameters are: ",grid.best_params_)
print("The best score is: ",grid.best_score_)
```

```python
# results of gridsearchcv
result = grid.cv_results_
table = pd.DataFrame.from_dict(result)
table
```

```python
# User inputs
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=float(input("Sex Male 0 or 1:"))
smoker_yes_input=float(input("Smoker yes 0 or 1:"))
```

```python
# Model prediction
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print('Future_Prediction = ',Future_Prediction)
```
